```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations from stooq.pl service
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df.info()
```

### Obliczanie nowych kolumn

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head(3)
```

```python
df['Average'] = (df.Open + df.Close) / 2.0
df.head()
```

```python
df.sort_index()
```

```python
df = df.sort_index()
```

```python
df.head()
```

```python
df[['Close']].shift(1)
```

```python
df['Close_shift'] = df.Close.shift(1)
```

```python
df.head()
```

```python
df.Close / df.Close_shift - 1
```

```python
df['Daily_change'] = df.Close / df.Close_shift - 1
df.head()
```

```python
df.Daily_change.min()
```

```python
df.Daily_change.max()
```

```python
import seaborn as sns
sns.set()
df.Daily_change.hist(bins=100)
```
