```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')
df = fetch_financial_data('META')
```

```python
df.head()
```

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2025)]
df_nov
```

```python
df_nov.to_csv('meta_nov.csv')
```
